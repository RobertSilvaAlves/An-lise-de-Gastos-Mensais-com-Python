# **Análise de Despesas Pessoais Mensais com Python**

# Contexto

Uma pessoa deseja entender melhor seus gastos mensais, a fim de tomar decisões financeiras mais conscientes e identificar oportunidades de redução ou redistribuição de despesas.

- **Introdução:**

O objetivo deste projeto é criar um controle financeiro simples, analisando os gastos mensais divididos por categoria e valor total por mês.

A análise permite compreender:

Quanto é gasto em cada mês

Quais categorias concentram os maiores gastos

Como os gastos se distribuem ao longo do período analisado

Essas informações auxiliam na tomada de decisões financeiras mais estratégicas.

-----

- **<font color = green>Estrutura dos Dados**

Os dados foram organizados em um dicionário, onde:

- Cada chave principal representa um mês

- Cada mês contém um conjunto de categorias de gastos

- Cada categoria possui um valor numérico associado

Estrutura conceitual:
```
Mês
 └── Categoria → Valor
```

Essa abordagem foi escolhida por facilitar:

- A organização hierárquica dos dados

- O acesso direto aos valores por mês e categoria

- A realização de cálculos e visualizações

Os valores totais mensais não fazem parte dos dados brutos, sendo calculados posteriormente a partir das categorias.

---

- **<font color=green>Dados Utilizados**

- - Para fins de simulação, foram considerados dois meses:

**Janeiro**

- Alimentação

- Transporte

- Lazer

- Moradia

**Fevereiro**

- Alimentação

- Transporte

- Lazer

- Moradia

Os valores foram definidos diretamente no código para manter o foco na análise dos dados, e não na entrada interativa.

---

- **<font color=green>Tecnologias Utilizadas**

- Python

- Matplotlib (visualização de dados)

- Google Colab (ambiente de desenvolvimento)

---

- **<font color=green>Análises Realizadas**

- **Total de Gastos por Mês**

- - Foi calculado o total de gastos para cada mês a partir da soma dos valores das categorias.

- - Essa análise permite identificar qual mês apresentou maior despesa total.

- **Categoria com Maior Gasto**

- - Para cada mês, foi identificada a categoria que apresentou o maior valor de gasto, permitindo compreender quais áreas têm maior impacto no orçamento mensal.

- **Comparação entre Meses**

- - Os valores totais mensais foram comparados para identificar variações entre os meses analisados.

---

- **<font color=green>Visualizações**

- **Total de Gastos por Mês**

- - Foi criado um gráfico de barras para comparar visualmente o total de gastos entre Janeiro e Fevereiro.

- - Esse gráfico facilita a identificação rápida do mês com maior despesa total.


- **Gastos por Categoria — Janeiro e Fevereiro**

- - Foram criados gráficos de barras individuais para cada mês, mostrando a distribuição dos gastos por categoria.

**Características dos gráficos:**

- Categorias ordenadas em ordem crescente de valor

- Valores exibidos diretamente sobre as barras

- Títulos e eixos bem definidos

Essas visualizações tornam a análise mais clara e intuitiva.

````python
import matplotlib
````

````python
dados = {
    'JANEIRO':{
        'Alimentação': 820,
        'Transporte': 420,
        'Lazer': 260,
        'Moradia': 1200
    },
    'FEVEREIRO':{
        'Alimentação': 760,
        'Transporte': 480,
        'Lazer': 180,
        'Moradia': 1200
    }
}

total_janeiro = sum(dados['JANEIRO'].values())
total_fevereiro = sum(dados['FEVEREIRO'].values())

categoria_maior_gasto_janeiro = max(dados['JANEIRO'], key=dados['JANEIRO'].get)
valor_maior_gasto_janeiro = max(dados['JANEIRO'].values())

categoria_maior_gasto_fevereiro = max(dados['FEVEREIRO'], key=dados['FEVEREIRO'].get)
valor_maior_gasto_fevereiro = max(dados['FEVEREIRO'].values())

print(f'O total de gastos de JANEIRO foi: {total_janeiro}\nO total de gastos de FEVEREIRO foi: {total_fevereiro}')
print(f'A categoria que teve maior gasto em JANEIRO foi: {categoria_maior_gasto_janeiro}')
print(f'A categoria que teve maior gasto em FEVEREIRO foi: {categoria_maior_gasto_fevereiro}')

if valor_maior_gasto_janeiro > valor_maior_gasto_fevereiro:
  print(f'O maior gasto foi em Janeiro na categoria {categoria_maior_gasto_janeiro}')

elif valor_maior_gasto_janeiro == valor_maior_gasto_fevereiro:
  print(f'Houve empate na categoria {categoria_maior_gasto_janeiro} de JANEIRO e na categoria {categoria_maior_gasto_fevereiro} de FEVEREIRO, com valor de {valor_maior_gasto_janeiro} ')

else:
  print(f'O maior gasto foi em Fevereiro na categoria {categoria_maior_gasto_fevereiro}')

if total_janeiro > total_fevereiro:
  print(f'O mês com mais gasto total foi Janeiro, com total de {total_janeiro}')
else:
  print(f'O mês com mais gasto total foi Fevereiro, com total de {total_fevereiro}')
````

````python
import matplotlib.pyplot as plt

mes = list(dados.keys())
valores_totais = [total_janeiro, total_fevereiro]


plt.bar(x = mes, height = valores_totais)

plt.title('Gastos Totais Por Mês')
plt.xlabel('Mês')
plt.ylabel('Valor Total')
````

````python
categorias = []
valores = []

for categoria, valor in dados['JANEIRO'].items():
  categorias.append(categoria)
  valores.append(valor)


plt.bar(x = categorias, height = valores)

plt.title('Gastos por Categoria — Janeiro')
plt.xlabel('Categoria')
plt.ylabel('Valores')
````

````python
#Transformndo os Gastos por Categoria — Janeiro em ordem crescente


lista_ordenada = zip(*sorted(zip(valores, categorias)))

valores, categorias = map(list, lista_ordenada)

bars = plt.bar(x = categorias, height = valores)

plt.bar_label(bars, fmt='%.1f', label_type='edge', size=10)

plt.title('Gastos por Categoria — Janeiro')
plt.xlabel('Categoria')
plt.ylabel('Valores')
````

````python
categorias = []
valores = []

for categoria, valor in dados['FEVEREIRO'].items():
  categorias.append(categoria)
  valores.append(valor)

#Transformndo os Gastos por Categoria — FEVEREIRO em ordem crescente

lista_ordenada = zip(*sorted(zip(valores, categorias)))

valores, categorias = map(list, lista_ordenada)

bars = plt.bar(x = categorias, height = valores)

plt.bar_label(bars, fmt='%.1f', label_type='edge', size=10)

plt.title('Gastos por Categoria — Fevereiro')
plt.xlabel('Categoria')
plt.ylabel('Valores')
````

# **<font color=green>Insights Obtidos**

- O mês de Janeiro apresentou maior gasto total em comparação com Fevereiro

- A categoria Moradia foi a que concentrou o maior gasto em ambos os meses

- Gastos com Lazer foram os menores entre as categorias analisadas
